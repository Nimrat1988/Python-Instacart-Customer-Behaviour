#6 - The team now wants to target different types of spenders in their marketing campaigns. This can be achieved by looking at the prices of the items people are buying. Create a spending flag for each user based on the average price across all their orders using the following criteria:
If the mean of the prices of products purchased by a user is lower than 10, then flag them as a “Low spender.”
If the mean of the prices of products purchased by a user is higher than or equal to 10, then flag them as a “High spender.”

```python
#Aggregating data with transform() function
df_ords_prods_merge['avg_product_price'] = df_ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
#To check results
df_ords_prods_merge[['user_id', 'avg_product_price']].head(15)
```

```python
#To create spending flag
df_ords_prods_merge.loc[df_ords_prods_merge['avg_product_price'] >= 10, 'spending_flag'] = 'High spender'
df_ords_prods_merge.loc[df_ords_prods_merge['avg_product_price'] < 10, 'spending_flag'] = 'Low spender'
```

```python
#To print frequency of spending flag
df_ords_prods_merge['spending_flag'].value_counts(dropna = False)
```

```python
#To check output of user_id, avg_product_price and spending flag
df_ords_prods_merge[['user_id', 'avg_product_price', 'spending_flag']].head(15)
```

#7 - In order to send relevant notifications to users within the app (for instance, asking users if they want to buy the same item again), the Instacart team wants you to determine frequent versus non-frequent customers. Create an order frequency flag that marks the regularity of a user’s ordering behavior according to the median in the “days_since_prior_order” column. The criteria for the flag should be as follows:
If the median of “days_since_prior_order” is higher than 20, then the customer should be labeled a “Non-frequent customer.”
If the median is higher than 10 and lower than or equal to 20, then the customer should be labeled a “Regular customer.”
If the median is lower than or equal to 10, then the customer should be labeled a “Frequent customer.”

```python
#Aggregating data with transform() function
df_ords_prods_merge['order_regularity_median'] = df_ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
#To check the output
df_ords_prods_merge[['user_id', 'days_since_prior_order', 'order_regularity_median']].head(15)
```

```python
#To create order frequency flag
df_ords_prods_merge.loc[df_ords_prods_merge['order_regularity_median'] > 20, 'order_frequency_flag'] = 'Non-frequent customer'
df_ords_prods_merge.loc[(df_ords_prods_merge['order_regularity_median'] <= 20) & (df_ords_prods_merge['order_regularity_median'] > 10), 'order_frequency_flag'] = 'Regular customer'
df_ords_prods_merge.loc[df_ords_prods_merge['order_regularity_median'] <= 10, 'order_frequency_flag'] = 'Frequent customer'
```

```python
#To print frequency of order_frequency_flag
df_ords_prods_merge['order_frequency_flag'].value_counts(dropna = False)
```

```python
#To check columns for the updated dataframe
df_ords_prods_merge[['user_id', 'days_since_prior_order', 'order_regularity_median', 'order_frequency_flag']].head(15)
```
