After importing the cleaned products CSV file, I noticed that the row count did not match the cleaned dataset. Because of this, I repeated the consistency checks to confirm the correct number of rows.

```python
#Remove Unnamed:0 coulumn
df_prods.drop(columns=['Unnamed: 0'], inplace=True)
```

```python
#Create new dataframe
df_prods_clean = df_prods[df_prods['product_name'].isnull() == False]
```

```python
#Finding duplicates
df_dups = df_prods_clean[df_prods_clean.duplicated()]
```

```python
#Create a dataframe that doesn't include duplicates
df_prods_clean_no_dups = df_prods_clean.drop_duplicates()
```

Suitable way to merge the order_products_combined dataframe with the products dataframe will be using product_id column.

```python
#Test the merge without overwriting
pd.merge(df_ords_cmbd, df_prods, on = 'product_id', indicator = 'match')
```

```python
#Drop themerge colum in the orders_products_combined dataframe
df_ords_cmbd = df_ords_cmbd.drop(['_merge'], axis = 1)
```

```python
#Merge updated dataframe
df_ords_prods_merge = df_prods_clean_no_dups.merge(df_ords_cmbd, on = 'product_id', indicator = True)
```
