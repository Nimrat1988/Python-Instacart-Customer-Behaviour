#Merging Instacart Data

```python
# Import libraries

import pandas as pd
import numpy as np
import os
```

```python
#Project folder path as a string
path = r'/Users/Nimrat/Documents/Instacart Basket Analysis'
```

```python
#Import the “order_product_prior.csv” data set into your notebook using the os library
df_ords_prior = pd.read_csv(os.path.join(path, 'Original data', 'orders_products_prior.csv'), index_col = False)
```

```python
#Check the output
df_ords_prior
```

```python
#Import the “order_checked.csv” data set into your notebook using the os library
df_ords = pd.read_csv(os.path.join(path, 'Prepared data', 'orders_checked.csv'), index_col = False)
```

```python
# Merge df_ords_prior and df_ords using order_id as a key
df_merged_large = df_ords.merge(df_ords_prior, on = 'order_id', indicator = True)
```
